### Import packages

```python
from requests import get
from requests.exceptions import RequestException
from contextlib import closing
from bs4 import BeautifulSoup
```

```python
# get webpage
def simple_get(url):
    
    try:
        with closing(get(url, stream=True)) as resp:
            if is_good_resp(resp):
                return resp.content
            else:
                pass
            
    except RequestException as e:
        log_error(f'Error during requsts to {url}: {str(e)}')
        pass
    
def is_good_resp(resp):
    
    content_type = resp.headers['Content-Type'].lower()
    return (resp.status_code == 200
            and content_type is not None
            and content_type.find('html') > -1)

def log_error(e):
    
    print(e)
    
```

```python
# search parameters Data Scientist jobs in Greater Los Angeles
raw_html = simple_get('https://www.linkedin.com/jobs/search/?alertAction=viewjobs&currentJobId=1579614147&geoId=90000049&keywords=data%20scientist&location=Los%20Angeles%20Metropolitan%20Area')
len(raw_html)
```

```python
# parse raw html
html = BeautifulSoup(raw_html, 'html.parser')
print(html.prettify())
```

```python
# find all job listing titles
titles = html.find_all(class_='result-card__title job-result-card__title')
print(len(titles))
#print(titles)
```
